```
import pickle
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
from keras.utils import plot_model 
```

**LSTM WITH W2V AND GLOVE**

```
file1 = open('LSTM+Word2Vec', 'rb')
history1 = pickle.load(file1)
file1.close()

file2 = open('LSTM+Glove','rb')
history2 = pickle.load(file2)
file2.close()

fig = plt.figure(figsize=(10,9))
ax = fig.add_subplot(111)
ax.plot(history1['val_acc'])
ax.plot(history2['val_acc'])


ymax1 = max(history1['val_acc'])
xmax1 = history1['val_acc'].index(ymax1)

ymax2 = max(history2['val_acc'])
xmax2 = history2['val_acc'].index(ymax2)


plt.plot(xmax1,ymax1,'ro')
plt.text(xmax1*1.01, ymax1*1.00 , '(%d, %.3f)'%(xmax1,ymax1), fontsize=11)

plt.plot(xmax2,ymax2,'ro')
plt.text(xmax2*0.87, ymax2*1 , '(%d, %.3f)'%(xmax2,ymax2), fontsize=11)


plt.title('Epoch(10) difference between W2V and Glove with LSTM', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
plt.xlabel('Epoch', fontsize=18)
plt.legend(['LSTM + GloVe', 'LSTM + W2V'], loc='upper left',fontsize=13)
plt.savefig("LSTM+W2V+GloVe.png")
```

**BIDORECTIONAL LSTM WITH W2V AND GLOVE**


```
file1 = open('Bi-LSTM+W2V', 'rb')
history1 = pickle.load(file1)
file1.close()

file2 = open('Bi-LSTM+Glove','rb')
history2 = pickle.load(file2)
file2.close()

fig = plt.figure(figsize=(10,9))
ax = fig.add_subplot(111)
ax.plot(history1['val_acc'])
ax.plot(history2['val_acc'])


ymax1 = max(history1['val_acc'])
xmax1 = history1['val_acc'].index(ymax1)

ymax2 = max(history2['val_acc'])
xmax2 = history2['val_acc'].index(ymax2)


plt.plot(xmax1,ymax1,'ro')
plt.text(xmax1*1.01, ymax1*1.00 , '(%d, %.3f)'%(xmax1,ymax1), fontsize=11)

plt.plot(xmax2,ymax2,'ro')
plt.text(xmax2*0.87, ymax2*1 , '(%d, %.3f)'%(xmax2,ymax2), fontsize=11)


plt.title('Epoch(10) difference between W2V and Glove with Bi-LSTM', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
plt.xlabel('Epoch', fontsize=18)
plt.legend(['Bi-LSTM + GloVe', 'Bi-LSTM + W2V'], loc='upper left',fontsize=13)
plt.savefig("Bi-LSTM+W2V+GloVe.png")
```

**BIDIRECTIONAL LSTM WITH RELU ACTIVATION**

```
import matplotlib.pyplot as plt
import h5py

file1 = open('Bi-LSTM+Glove+Activation', 'rb')
history = pickle.load(file1)
file1.close()




# Plot training & validation accuracy values
plt.plot(history['acc'])
plt.plot(history['val_acc'])
plt.title('Model accuracy for Bi-directional LSTM with Relu Activation')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
plt.savefig("Bi-LSTM+Activation_Acc.png")


# Plot training & validation loss values
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Model loss for Bi-directional LSTM with Relu Activation')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
plt.savefig("Bi-LSTM+Activation_Loss.png")
```
